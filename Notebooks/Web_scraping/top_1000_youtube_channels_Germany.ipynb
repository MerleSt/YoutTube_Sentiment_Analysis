```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://tz.youtubers.me/germany/all/top-1000-most-subscribed-youtube-channels-in-germany/en'
r = requests.get(url)
soup = BeautifulSoup(r.content, 'html.parser')
```

```python
print(soup.prettify())
```

```python
# Locate the table with class 'top-chart-box'
table = soup.find('table', class_='top-charts')
```

```python
print(table)
```

```python
df = pd.read_html(str(table))[0]
```
